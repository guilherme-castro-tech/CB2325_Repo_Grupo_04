```python
# Exemplo de função:

def criar_lista(a, b: int=1):
    
    """
    Cria uma lista de b elementos 'a' utilizando o for.

    Parâmetros:
    a : Elementos que vão preencher a lista;
    b (Inteiro positivo): Tamanho da lista.

    Retorna: Uma lista de tamanho b
    """

    if b < 0:
        raise ValueError("O tamanho da lista precisa ser um número positivo")

    l = []
    for i in range(b):
        l.append(a)

    return l


if __name__ == "__main__":

    # Testes:
    print("Lista de oito números 5:")
    print(criar_lista(5, 8))
    print()
    print("Lista de dois floats 0.543532:")
    print(criar_lista(0.543532, 2))
    print()
    print("Lista de quadro listas ['IMPA Tech', 2025]:")
    print(criar_lista(["IMPA Tech", 2025], 4))
```

```python
# Exemplo de função:

def criar_pontos(funcao, intervalo: list):
    
    """
    Cria uma lista de pontos segundo uma determinada função real.   

    Parâmetros:     
    funcao : Função real para ser avaliada no intervalo;  
    intervalo (Lista): Intervalo em que os pontos vão ser gerados.

    Retorna: Uma lista de pontos da função desejada
    """

    xcoords = list(range(intervalo[0], intervalo[1]+1))

    ycoords = list(funcao(x) for x in xcoords)

    return list(zip(xcoords, ycoords))


if __name__ == "__main__":

    # Testes:
    print("Função x**2 + 2:")
    f = lambda x: x**2 + 1
    print(criar_pontos(f, [0,12]))
    print()
    print("Função 3*x**2 + 4:")
    f = lambda x: 2*x**2 + 4
    print(criar_pontos(f, [4,16]))
    print()
    print("Função 2*x:")
    f = lambda x: 2*x
    print(criar_pontos(f, [0,10]))
```

```python
# Padrão de plotagem

def padrao_plotagem(intervalo: list, funcao1, pontos1: list, funcao2, pontos2: list):

    """
    Utiliza os padrões de plotagem definidos pelo grupo para elaborar um gráfico "padrão".

    Parâmetros:     
    intervalo (Lista): Intervalo das funções;   
    funcao1 : 1ª função para ser avaliada no intervalo;  
    pontos1 (Lista): Lista de pontos (x,y) da 1ª função;    
    funcao2 : 2ª função para ser avaliada no intervalo;  
    pontos1 (Lista): Lista de pontos (x,y) da 2ª função.

    Retorna: Um gráfico que segue os padrões definidos nessa função
    """

    # Paleta de cores padrão definida pelo grupo:
    Paleta = ["#084b83", "#680e4b", "#c42021", "#edae49"]   # -> Ordenadas por prioridade

    # Dar preferência para plotagem de gráficos sempre em azul, se houver necessidade de plotar mais de uma função, seguir a ordem de prioridade
    # Organizar as plotagens de forma que sempre exista um contraste significativo entre os elementos e uma boa visualização dos dados
    # Em casos que exigem a plotagem de mais de 4 elementos, selecione uma cor que cumpra esses requisitos e notifique 

    conjunto = [pontos1, pontos2]
    xcoords, ycoords = [[],[]], [[],[]]
    aux = 0
    for pontos in conjunto:
        aux += 1
        for i,j in pontos:
            xcoords[aux-1].append(i)
            ycoords[aux-1].append(j)
    
    fig, ax = plt.subplots()
    X = np.linspace(min(intervalo), max(intervalo), 200)
    Y = funcao1(X)
    Y_2 = funcao2(X)

    ax.axhline(y=0, color="grey") # Caso seja necessário plotar algum dos eixos, utilizar a cor padrão 'grey'

    # Os gráficos são plotados segundo a ordem de prioridade

    ax.plot(X, Y, color=Paleta[0], label="Função 1")  # Gráfico prioritariamente em azul
    ax.plot(X, Y_2, color=Paleta[1], label="Função 2") # Gráfico 2 seguindo a ordem de prioridade

    # Os pontos devem ser plotados acima dos demais elementos para evitar sobreposições indesejadas

    ax.plot(xcoords[0], ycoords[0], "o", color=Paleta[2], label="Pontos da função 1")
    ax.plot(xcoords[1], ycoords[1], "o", color=Paleta[3], label="Pontos da função 2") 

    plt.title("Plotagem 'padrão'")    # Título inicia com letra maiúscula
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo Y')
    
    plt.grid(True)  # Grid ativado por padrão, mas pode ser desativado em caso de poluição visual do gráfico
    
    plt.legend() # Itens de legenda começando sempre com letra maiúscula

    plt.show()
```

```python
f = lambda x: x**2

pf = criar_pontos(f, [0,10])

g = lambda x: -x**2

pg = criar_pontos(g, [0,10])

padrao_plotagem([0,10], f, pf, g, pg)
```
